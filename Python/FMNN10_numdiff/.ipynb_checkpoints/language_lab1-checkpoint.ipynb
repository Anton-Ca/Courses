#### Writing a tokenizer 

Write a Unicode regular expression to find words defined as sequences of letters.

```python
# Write your regex here
regex = r"[A-Za-z|å|Å|ö|Ö|ä|Ä|é|-]+"
```

```python
re.findall(regex, 'En gång hade de på Mårbacka en barnpiga, som hette Back-Kajsa')
```

Using `regex`, write `tokenize(text)` function to tokenize a text. Return their positions.

```python
# Write your code here
def tokenize(text): 
    tokens = []
    regex = r"[A-Za-z|å|Å|ö|Ö|ä|Ä|é|-]+"
    words=re.findall(regex, text)
    words= set(words)
    for word in words:
            it = re.finditer(r"\b"+word+r"\b",text)
            for match in it:
                if [match.group(0).lower(), match.start()] not in tokens:
                    tokens.append([match.group(0).lower(), match.start()])
    return tokens
```

```python
tokens = tokenize('En gång hade de på Mårbacka en barnpiga, som hette Back-Kajsa.')
list(tokens)
```

#### Extracting indices

Write a `text_to_idx(words)` function to extract the indices from the list of tokens (words). Return a dictionary, where the keys will be the tokens (words), and the values a list of positions.

```python
# Write your code here
def text_to_idx(tokens): 
    token_positions = {}
    for token in tokens:
        if token[0] in token_positions and token[1] not in token_positions[token[0]]:
            token_positions[token[0]].append(token[1])
        else: 
            token_positions[token[0]] = [token[1]]
    return token_positions
```

```python
tokens = tokenize('En gång hade de på Mårbacka en barnpiga, som hette Back-Kajsa.'.lower().strip())
text_to_idx(tokens)
```

### Reading the content of a folder

Write a `get_files(dir, suffix)` function that reads all the files in a folder with a specific `suffix` (txt). You will need the Python `os` package, see <a href="https://docs.python.org/3/library/os.html">https://docs.python.org/3/library/os.html</a>. You will return the file names in a list.

You can reuse this function:

```python
def get_files(dir, suffix):
    """
    Returns all the files in a folder ending with suffix
    :param dir:
    :param suffix:
    :return: the list of file names
    """
    files = []
    for file in os.listdir(dir):
        if file.endswith(suffix):
            files.append(file)
    return files
```

```python
# write your code here
def pickle_idx(filename):
    with open('Selma/'+filename, 'r') as _file_:
        text = _file_.read().replace('\n', ' ') 
        tokens = tokenize(text.lower().strip())
        idx = text_to_idx(list(tokens))
    
    with open(filename+'_idx.pickle', 'wb') as outfile:
        pickle.dump(idx, outfile)

def open_all_idx(files): #öppnar alla idx-filer och lägger i en dict
    all_idx = {}
    for filename in files: 
        with open(filename+'_idx.pickle', 'rb') as infile:
            all_idx[filename] = pickle.load(infile)
    return all_idx

def create_master_idx(all_idx): 
    master_idx = {}
    for filename in all_idx: 
        for word in all_idx[filename]:
            if word in master_idx:
                master_idx[word][filename]= all_idx[filename][word]
            else: 
                master_idx[word] = {filename: all_idx[filename][word]}
    return master_idx  

#for file in files: 
#    pickle_idx(file)
all_idx = open_all_idx(files)
master_index = create_master_idx(all_idx)  
```

### Representing Documents with tf-idf

Once you have created the index, you will represent each document in your corpus as a dictionary. The keys of these dictionaries will be the words and you will define the value of a word with the tf-idf metric: 
1. Read the description of the tf-idf measure on Wikipedia (<a href="https://en.wikipedia.org/wiki/Tf%E2%80%93idf">https://en.wikipedia.org/wiki/Tf-idf</a>)
2. After reading the description, you probably realized that there are multiple definitions of tf-idf. In this assignment, 
 * Tf will be the relative frequency of the term in the document and 
 * idf, the logarithm base 10 of the inverse document frequency.

Conceptually, the tf-idf representation is a vector. In your program, you will keep this idea and use all the words in the corpus as keys: Each dictionary will include all the words of the corpus as keys. The value of the key is then possibly 0, meaning that the word in not in the document or is in all the documents as for the word `nils` in `gosta.tx`. 

As further work, you may think of optimizing this part.

```python
# Write your code here
def sum_words(master_index):
    result = {}
    for word in master_index: 
        for file in master_index[word]:
            if file in result:
                result[file] += len(master_index[word][file])
            else: 
                result[file] = len(master_index[word][file])
    return result

def term_frequency(term, filename, master_index): # not relative. Normalized in create_tfidf
    if filename in master_index[term]:
        return len(master_index[term][filename])
    else:
        return 0
    
def inverse_document_frequency(term, N, master_index): # N = number of corpus files
    if term in master_index:
        d = len(master_index[term]) # number of documents where the term appears. 
    else:
        d = 0
    return math.log(N/d,(10))

def tf_idf(term, filename, N, master_index): 
    return term_frequency(term, filename, master_index)*inverse_document_frequency(term, N, master_index)

def create_tfidf(master_index, corpus_files): 
    sum_ = sum_words(master_index) # total number of words in each file. type = dict. 
    tfidf = {}
    for word in master_index: 
        for filename in corpus_files: 
            if filename in tfidf:
                tfidf[filename][word] = tf_idf(word, filename, len(corpus_files), master_index)/sum_[filename]
            else:
                tfidf[filename]={word : tf_idf(word, filename, len(corpus_files), master_index)/sum_[filename]}
    return tfidf

tfidf  = create_tfidf(master_index, files)
```

### Comparing Documents

Using the cosine similarity, compare all the pairs of documents with their tf-idf representation and present your results in a table. You will include this table in your report.

#### Cosine similarity

Write a function computing the cosine similarity between two documents

```python
# Write your code here
def cos_similarity(vec1, vec2): 
    return np.dot(vec1, vec2)/((np.linalg.norm(vec1)*np.linalg.norm(vec2)))


```

#### Similarity matrix

Compute the similarity matrix between the documents of the corpus. While computing the similarities, you will record the two most similar documents that you will call `most_sim_doc1` and `most_sim_doc2`.

```python
# Write your code here
def get_vectors(tfidf):
    vectors = {} 
    for filename in tfidf:
        vectors[filename]=tfidf[filename].values()
        
    return vectors

vectors = get_vectors(tfidf)

sim_matrix = [] 

for file1 in files:
    sim_row = []
    for file2 in files:
        sim = cos_similarity(list(vectors[file1]),list(vectors[file2]))
        sim_row.append(sim)
    sim_matrix.append(sim_row)

sim_matrix=np.array(sim_matrix)
np.fill_diagonal(sim_matrix, 0)
print(sim_matrix)
max_similarity = np.amax(sim_matrix)
ind1, ind2 = np.unravel_index(np.argmax(sim_matrix, axis=None), sim_matrix.shape)

most_sim_doc1 = files[ind1]
most_sim_doc2 = files[ind2]



```

Give the name of the two novels that are the most similar.

```python
print("Most similar:", most_sim_doc1, most_sim_doc2, "Similarity:", max_similarity)
```
